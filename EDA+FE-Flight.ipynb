```python
## Feature Engineering
# Fetching the dates
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
# Fetching the Month
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
# Fetching the Year
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
# Converting to Integer Dtype
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
# Drop Date of Journey Column
df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# To get only Hour and Min and not any other data we will apply Lambda Fn
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
# Separating the hour and mins
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
# Converting to Integer Dtype
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Min']=df['Arrival_Min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# Fetching same for Departure time
df['Departure_hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
# Converting into INT Dtype
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# Categorical Data FE
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'].unique()
```

```python
# Replacing Nan Value with 1 Stop
# Converting this into numerical/ordinal labels data to correlate with prices
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df[df['Total_Stops'].isnull()] # No Null Values
```

```python
df.head(2)
```

```python
# Drop "Route" as we have already 'Source' & 'Destination'
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# Converting Duration separately into Numerical Data
# Also filling the Na values with zeros
# Here we are using extract feature for Duration Hour to extract Duration hour as NA Values are there in Mins
df['Duration_Hr']=df['Duration'].str.extract(r'(\d+)h').fillna(0).astype(int)
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0).astype(int)
```

```python
df.drop('Duration',axis=1,inplace=True)
```
